Creating a condn column from more than 2 choices: np.select()

```python
import numpy as np
```

```python
#create an array based on multiple price tiers (+2 choice)
#create   "conditions"and "values" variables
conditions=[
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>1000) & (df_laptops['Price_euros']<=2000),
    (df_laptops['Price_euros']>500) & (df_laptops['Price_euros']<=1000)
]
values=['Too Expensive','Expensive','Afordable','Cheap']
```

```python
#Set into new column
df_laptops['Price_tier_multi']=np.select(conditions,values, default='Very Cheap')
df_laptops.head()
```

```python
df_laptops['Price_tier_multi'].value_counts()
```

```python
#create an array based on multiple screen size tiers (+2 choice)
#Too Big>16,Big>14 small>12 , Too Small<12
conditions=[
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    df_laptops['Inches']<=12

]
values=['Too Big','Big','Small',"Too Small"]
#create "condn" and values "variables"
```

```python
#set into column
df_laptops['Screen_size']=np.select(conditions,values,default='Medium')
df_laptops.head()
```

```python
#count values
df_laptops['Screen_size'].value_counts()
```

Representing it by Graph

```python
# prompt: Using dataframe df_laptops: bar graph

import altair as alt
# Create a bar chart using Altair
chart = alt.Chart(df_laptops).mark_bar().encode(
    # Encode Company on the x-axis
    x='Company',
    # Encode Price_euros on the y-axis
    y='Price_euros'
).properties(
    # Set the title of the chart
    title='Average Price by Company'
)
# Display the chart
chart
```

```python


# Group data by Company and count the occurrences
company_counts = df_laptops['Company'].value_counts().reset_index()
company_counts.columns = ['Company', 'Count']

# Create the pie chart using Altair
pie_chart = alt.Chart(company_counts).mark_arc().encode(
    # Encode the count for the arc length
    theta=alt.Theta(field="Count", type="quantitative"),
    # Encode the Company for color
    color=alt.Color(field="Company", type="nominal"),
    # Add tooltips to show Company and Count
    tooltip=['Company', 'Count']
).properties(
    # Set the title of the chart
    title='Distribution of Laptops by Company'
)

# Display the chart
pie_chart
```

```python


# Assuming 'df_laptops' is already loaded and available from the preceding code

# Group by 'Company' and calculate the sum of sales (assuming 'Price_euros' represents sales or revenue here)
company_sales = df_laptops.groupby('Company')['Price_euros'].sum().reset_index()

# Create the line chart using Altair
line_chart = alt.Chart(company_sales).mark_line(point=True).encode(
    # Encode Company on the x-axis
    x=alt.X('Company', sort='y'), # Sort by sales for better visualization
    # Encode the sum of Price_euros on the y-axis
    y='Price_euros'
).properties(
    # Set the title of the chart
    title='Total Sales (Price) by Company'
)

# Display the chart
line_chart
```
